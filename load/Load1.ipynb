```python
import json
import pandas as pd
from ast import literal_eval
import glob
import csv
import copy
```

```python
data = []

files = sorted(glob.glob('../fsm-data/*', recursive=False))

for single_file in files:
    with open(single_file, 'r') as f:
        json_file = json.load(f)
        new_string = json.dumps(json_file, indent = 2)
        #print(new_string)
        data.append(json_file)
```

```python
c = []
for single_file in data:
    temp = single_file['states']
    d = []
    #print('------------------------------')
    for i in temp:
        att = list(i.keys())
        for key in att:
            if (key == 'displayText'):
                s = ''
                text = list(i[key].keys())
                for j in text:
                    s += i[key][j]
                if (not (s == '')):
                    d.append(s)
    c.append(d)
```

```python
import nltk
nltk.download('all')
from nltk import tokenize
```

The following code is to break down the big paragraph into sentences, and then apply [CLS] and [SEP] into start and end of each sentence (for BERT)

```python
sentences = []
for i in c:
    sentences_i = []
    for j in i:
        p = nltk.sent_tokenize(j) 
        sentences_i.append(p)
    sentences.append(sentences_i)
#print(sentences[1])
```

```python
sentences_tuned = []
index = 0
for i in sentences:
    sentences_tuned.append([])
    for j in i:
        for k in j:
            temp = "[CLS] " + k + " [SEP]"
            sentences_tuned[index].append(temp)
    index += 1
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
tokenized_paras = []
for para in sentences_tuned:
    tokenized_texts = [tokenizer.tokenize(sent) for sent in para]
    tokenized_paras.append(tokenized_texts)
print(tokenized_paras[0])
        
```
